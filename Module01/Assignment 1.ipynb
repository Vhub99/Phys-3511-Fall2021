# Problem 1
## Stirling's Approximation

```python
for n in range(2,20):
    v1 = np.math.log(np.math.factorial(n)) #ln(n!)
    v2 = n * np.math.log(n) - n #stirling's_approximation
    
    print(v1, v2)
```

```python
for n in range(2,91):
    v1 = np.math.log(math.factorial(n))
    v2 = n * np.math.log(n) - n
    e = (v1 - v2) / v1
    if e < 0.01:
        print(round(v1,3), round(v2,3), round(e,3))
```
